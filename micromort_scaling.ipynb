# How do micromorts add up?

```python
import numpy as np
import matplotlib.pyplot as plt
```

Micromorts are a convenient way to measure the dealiness of human activity.
They are defined as the amount of activity resulting in a 1:1 000 000 chance of dying.

These low probablities can be added up, due to the taylor expansion

$\begin{align}
p_{not\_dead} &= (1 - p)^N \\
              &=  1 - N \cdot (1 -0) \cdot p + O(p^2)\\
              &\approx 1 - Np
\end{align}$

But this is clearly not valid for $N\approx \frac{1}{p}$, as $p_{not\_dead} = 0$ in this case.

So, lets simulate some numbers!

```python
n_long = np.logspace(1, 6.5, 100)
p_not_death = np.power(1 - 10**-6, n_long) 
plt.plot(n_long, p_not_death, label='exact')

n = np.logspace(1, 6, 100)
taylored_not_death = 1 - n * 10**-6
plt.plot(n, taylored_not_death, label='taylored')
plt.figure()
plt.show()
```

As we can see, the exact probablity to zero asymptotically, and is about zero at $N\approx \frac{3}{p}$, while the taylored formula is exactly zero at $N = \frac{1}{p}$.

But for which quantities of N can we use the first-order taylored formula?

```python
n = np.logspace(1, 5.7, 1000)
p_not_death = np.power(1 - 10**-6, n) 
plt.plot(n, p_not_death, label='exact')

taylored_not_death = 1 - n * 10**-6
plt.plot(n, taylored_not_death, label='taylored')
plt.figure()
plt.show()
```

It's probably safe for $N<\frac{1}{p}/10$, and roughly correct for $N<\frac{1}{p}/5$, 100 000 and 500 000 for the micromorts in this case.
